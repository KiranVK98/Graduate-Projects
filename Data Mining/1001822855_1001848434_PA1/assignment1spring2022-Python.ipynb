# Task 2: Aggregation & Filtering & Rank
In this task, we will perform some very high level aggregation and filtering operations. Then, we will apply ranking on the results for some tasks. Pandas has a convenient and powerful syntax for aggregation, filtering, and ranking. DO NOT write a for loop. Pandas has built-in functions for all tasks. 

```python
#Task 2-a: Find out the sum of Captial Gain for people with education level as Bachelors and HS-Grad.

sum_capital_gain_bachelors = sum(df_census.loc[df_census['education'] == ' Bachelors']['capital-gain'])
sum_capital_gain_HS_GRAD = sum(df_census.loc[df_census['education'] == ' HS-grad']['capital-gain'])
print ("Task 2-a: The sum of capital gain for education level as bachelors is %s and as HS-Grad is %s" 
    % (sum_capital_gain_bachelors, sum_capital_gain_HS_GRAD))

###############begin your code here

#Task 2-b: Find out the total number of people surveyed in months may, october and december.
#Create a new column for 'Survey_Month' by using 'Date' column
#write the code for extracting the month from the date column here

df_census['Date'] = pd.to_datetime(df_census['Date'])
df_census['Survey_Month'] = df_census['Date'].dt.strftime('%B')
num_surveys_may = df_census.loc[pd.to_datetime(df_census['Date']).dt.month == 5].shape[0]
num_surveys_october = df_census.loc[pd.to_datetime(df_census['Date']).dt.month == 10].shape[0]
num_surveys_december = df_census.loc[pd.to_datetime(df_census['Date']).dt.month == 12].shape[0]
print ("\n\nTask 2-b: The total number of surveys in may is %s, in october is %s, and in december is %s" 
        % (num_surveys_may, num_surveys_october, num_surveys_december))

###############send you code here

#Task 2-c: Let us now use multiple filtering criteria
# Find out the total number of surveys in september and november with workclass as private and age less than 50.

num_surveys_september = len(df_census.loc[(df_census['Survey_Month'] == 'September') & (df_census['WorkClass'] == ' Private') & (df_census['Age'] < 50)])
num_surveys_november = len(df_census.loc[(df_census['Survey_Month'] == 'November') & (df_census['WorkClass'] == ' Private') & (df_census['Age'] < 50)])
print ("\n\nTask 2-c: The total number of surveys that meet the given conditions in september is %s and in november is %s" 
       % (num_surveys_september, num_surveys_november))


#Task 2-d: Find out 3 least surveyed education categories, print their names and corresponding number of surveys for 
#periods January-June and July-December.

list_1 = ['January','February','March','April','May','June']
list_2 = ['July','August','September','October','November','December']
df_2d = df_census.groupby(["Survey_Month","education"]).count().nsmallest(50, 'Date').reset_index().rename(columns={"Date": "Count"})
p1_2d = df_2d[df_2d['Survey_Month'].isin(list_1)]
p2_2d = df_2d[df_2d['Survey_Month'].isin(list_2)]
top3_least_surveyed_Jan_June = p1_2d.drop_duplicates(["education"]).head(3)[["education","Count"]]
top3_least_surveyed_July_December = p2_2d.drop_duplicates(["education"]).head(3)[["education","Count"]]
print ("\n\nTask 2-d: \nThe top 3 least surveyed education categories in January-June: \n%s \n\nThe top 3 least surveyed education categories in July-December: \n%s" 
                     % (top3_least_surveyed_Jan_June,top3_least_surveyed_July_December))

#Task 2-e: Find out top 5 native-countries besides United-States, print their names and number of surveys belonging 
#to each.
df_2e = df_census.groupby(["native-country"]).count().reset_index().rename(columns={"Date": "Count"})
p1_2e = df_2e.sort_values(by=['Count'], ascending=False).head(6)[["native-country","Count"]]
top5_most_surveyed_native_countries = p1_2e.iloc[1: , :]
print ("\n\nTask 2-e: \nThe top 5 most surveyed native countries : \n%s" 
                     % (top5_most_surveyed_native_countries))


#Task 2-f: Find out Top-5 native-countries with the most number of samples belonging to class >50K
df_2f = df_census.groupby(["native-country","class"]).count().reset_index().rename(columns={"Date": "Count"})
p1_2f = df_2f[df_2f['class'].isin([' >50K'])]
top5_native_countries = p1_2f.sort_values(by=['Count'], ascending=False).head(5)[["native-country","class","Count"]]
print ("\n\nTask 2-f: \nThe top 5 native countries with the most number of surveys with class >50K: \n%s" 
                     % (top5_native_countries))
```
